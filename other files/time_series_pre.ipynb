python np array由零開始
以下為將一行的內容六個為單位做整理

```python
time_series=np.zeros((len(y)-5,6))
time_series.shape
i=0
j=0

np.set_printoptions(threshold=np.inf)


for i in range(0,len(y)-5):
    for j in range(0,6):
        time_series[i][j]=y[i+j]
        j+=1
    i+=1
    j=0

time_series.shape
#print(time_series)
```

將time series存成excel
需要將二維陣列存成Dataframe

```python

store=pd.DataFrame(time_series,columns=["1","2","3","4","5","6"])
store.to_excel("time_series_of_total_value.xlsx",index=False)
```

```python
pre=pd.read_excel("time_series_of_total_value.xlsx")
pre.head(5)
```

```python
from sklearn.model_selection import train_test_split
x=pre.drop("6",axis=1)
y=pre["6"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
y_train=y_train.astype(int)
y_test=y_test.astype(int)
print(x_test[0:5])
print(x_train[0:5])
print(y[0:5])
```

```python
#引入scalar這個函數
from sklearn.preprocessing import StandardScaler
#使用scalar將StandardScaler()存入 用以將數據標準化 均值為零 標準差為1
#以下是標準寫法 將test 及 train 轉為標準化
scalar=StandardScaler()
x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
print(x_train_scaled.shape)
print(x_train[0:5])
print(x_train_scaled[0:5])
```

K鄰近

```python
#它是K最近鄰（K-Nearest Neighbors，KNN）分類算法的其中一種實現。KNN是一種監督學習算法，通常用於分類任務，根據輸入數據點在特徵空間中的鄰近點來進行分類。
from sklearn.neighbors import KNeighborsClassifier
#classification_report函式用於生成分類任務的報告，其中包括精確度、召回率、F1分數等評估指標，以評估模型的性能。
#accuracy_score函式用於計算分類的準確度，即模型正確分類的樣本比例。通常，這是一個最常用的分類性能指標之一。
from sklearn.metrics import classification_report,accuracy_score
#設定k值 會考慮附近多少個值去做定義 k小:噪聲大 k大:無法捕獲數據細節
knn= KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_scaled,y_train)
#將x_test之標準化值投入以上模型進行預測 並存入y_pred_knn中
y_pred_knn=knn.predict(x_test_scaled)
#比對預測及實際值
accuracy_knn=accuracy_score(y_test,y_pred_knn)
print(accuracy_knn)
#精確度（Precision）:指所有預測正確的數據中 實際正確的比率 實際正確/(實際正確+錯誤正確(假陽性)) 忽略對負例的預測
#召回率（Recall）：指所有原本正確值中 被分為正確值 分為正確值/實際正確值 忽略假陽性
#F1分數：F1分數是精確度和召回率的調和平均值，用於綜合考慮精確度和召回率。它適用於不平衡的數據集。
#支持數（Support）：每個類別的樣本數。
#加權平均：報告還包括宏平均（macro average）和加權平均（weighted average）的評估指標，用於全局性能總結。
classification_report_knn=classification_report(y_test,y_pred_knn)
print(classification_report_knn)
```

決策樹

```python
#引入決策樹分類器
from sklearn.tree import DecisionTreeClassifier
#42為某種參數
dt=DecisionTreeClassifier(random_state=42)
dt.fit(x_train_scaled,y_train)
y_pred_dt=dt.predict(x_test_scaled)

accuracy_dt=accuracy_score(y_test,y_pred_dt)
classification_report_dt=classification_report(y_test,y_pred_dt)

print(accuracy_dt)
print(classification_report_dt)
```

隨機森林

```python
from sklearn.ensemble import RandomForestClassifier
#42同樣為用以確保隨機性可重建 n_estimators表決策樹的數量
rf=RandomForestClassifier(random_state=42,n_estimators=100)

rf.fit(x_train_scaled, y_train)

y_pred_rf=rf.predict(x_test_scaled)

accuracy_rf=accuracy_score(y_test,y_pred_rf)
classification_report_rf=classification_report(y_test,y_pred_rf)

print(accuracy_rf)
print(classification_report_rf)
```
